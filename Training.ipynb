```python
import torch 
from noise2noise.datasets import load_dataset
from noise2noise.noise2noise import Noise2Noise
```

```python
class Args:
    train_source_dir = 'noise2noise/data/train/source'
    train_target_dir = 'noise2noise/data/train/target'
    valid_source_dir = 'noise2noise/data/valid/source'
    valid_target_dir = 'noise2noise/data/valid/target'
    ckpt_save_path = 'ckpts'
    ckpt_overwrite = True
    report_interval = 7
    train_size = None
    valid_size = None
    learning_rate = 1e-3
    adam = [0.9, 0.99, 1e-08]
    batch_size = 7
    nb_epochs = 30
    loss = 'l2'
    cuda = torch.cuda.is_available()
    plot_stats = True
    noise_type = 'gaussian'
    noise_param = 100
    seed = None
    crop_size = 128
    clean_targets = False
    no_crop = True
    add_noise = False

params = Args()
```

```python
train_loader = load_dataset(
    source_dir=params.train_source_dir,
    target_dir=params.train_target_dir,
    redux=params.train_size,
    params=params,
    shuffled=True,
    add_noise=params.add_noise
)

valid_loader = load_dataset(
    source_dir=params.valid_source_dir,
    target_dir=params.valid_target_dir,
    redux=params.valid_size,
    params=params,
    shuffled=False,
    add_noise=params.add_noise
)
```

```python
n2n = Noise2Noise(params, trainable=True)
```

```python
n2n.train(train_loader, valid_loader)
```
